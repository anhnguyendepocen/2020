## Euler's method

Let's write a function called `euler` which takes input parameters `f`, `t` and `y0` and returns the array of `y` values generated by Euler's method applied to $y' = f(t,y)$, $y(t_0)=y_0$.

```python
def euler(f,t,y0):
    y = np.zeros(t.shape)
    y[0] = y0
    for n in range(0,len(t) - 1):
        y[n + 1] = y[n] + f(t[n],y[n])*(t[n + 1] - t[n])
    return y
```

Test our function on an equation where we know the exact solution.

```python
def f(t,y):
    return -y

N = 50; a = 0; b = 2;
h = (b - a)/N
t = np.linspace(a,b,N + 1)
y0 = 1

y = euler(f,t,y0)
plt.plot(t,y,'.-')
plt.plot(t,np.exp(-t))
plt.legend(['Euler', 'Exact'])
plt.show()
```

Now let's approximate a weird equation: $y' = \sin(y)^t$

```python
def f(t,y):
    return (np.sin(y))**t

N = 100; a = 0; b = 4;
h = (b - a)/N
t = np.linspace(a,b,N + 1)

for y0 in np.linspace(0,1,11):
    y = euler(f,t,y0)
    plt.plot(t,y,'b-')
plt.show()
```
